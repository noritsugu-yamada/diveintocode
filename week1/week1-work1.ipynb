## 紙を43回おると厚みは月に届くか検証！

####  1.べき乗の算術演算を利用

```python
"""
紙を43回折った時の厚さを計算するコード
"""
# t0をthicknessとする
thickness = 0.00008

# 回数をnumとする
num = 43

# totalの厚みをfolded_thicknessとする
folded_thickness = thickness*2**num


print('厚さ: {:.2f}キロメートル'.format(folded_thickness/1000))
```

#### 2.for文を利用

```python
# for文の場合
thickness = 0.00008
num = 43
folded_thickness = thickness
for _ in range(1, num+1):
    folded_thickness *= 2
print('厚さ: {:.2f}キロメートル'.format(folded_thickness/1000))
```

月までの距離はレーザーによる測定で384,403.9kmとされている。

```python
moon_length = 384403.9
delta = folded_thickness/1000 - moon_length
print(delta)
```

delta>0なので月に届くと言える

#### 3.計算速度を比較

```python
import time
start = time.time()
####
# べき乗を使う場合
thickness = 0.00008
num = 43
folded_thickness = thickness*2**num
print('厚さ: {:.2f}キロメートル'.format(folded_thickness/1000))
####
elapsed_time = time.time() - start
print('tiem: {}[s]'.format(elapsed_time))
```

```python
import time
start_time = time.time()
####
# for文を使う場合
num = 43
thickness = 0.00008
for _ in range(1, num+1):
    thickness *= 2
folded_thickness = thickness
print('厚さ: {:.2f}キロメートル'.format(folded_thickness/1000))
elapsed_time = time.time() - start_time
print('time: {}[s]'.format(elapsed_time))
```

計算時間について

for文を使った場合べき乗計算に比べて時間がかかることがわかる
一般に実行に時間がかかることは望ましくないことなので
この場合べき乗計算の方が優れていると言える.

#### 4.可視化

```python
thickness  = 0.00008
folded_thickness = thickness
num = 43
folded_thickness_list  = []
folded_thickness_list.append(folded_thickness)
for _ in range(1,num+1):
    folded_thickness *= 2
    folded_thickness_list.append(folded_thickness)
    
print(len(folded_thickness_list))
```

```python
"""
グラフを表示する。タイトルと軸ラベル名付き。
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list)
```

```python
"""
グラフのy軸を対数に
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.yscale('log')
plt.plot(folded_thickness_list)
```

```python
"""
赤い線のグラフ
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list, color='red')
```

```python
"""
背景を黒、色を黄色、線のスタイルをdash-dot、線の太さを太くする
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use(['dark_background'])
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list, color='y', linestyle='-.', linewidth=3)
```
