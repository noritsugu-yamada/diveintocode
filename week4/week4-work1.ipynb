```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches

def decision_region(X_train, y_train, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は学習用データである。

    Parameters
    ----------------
    X_train : ndarray, shape(n_samples, 2)
        学習用データの特徴量
    y_train : ndarray, shape(n_samples,)
        学習用データの正解値
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X_train[:,0])-0.5, np.max(X_train[:,0])+0.5, step), np.arange(np.min(X_train[:,1])-0.5, np.max(X_train[:,1])+0.5, step))
    print(mesh_f0.shape, mesh_f1.shape)
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    print(mesh.shape)
    pred = model.predict(mesh).reshape(mesh_f0.shape)
    

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y_train)):
        plt.scatter(X_train[y_train==target][:, 0], X_train[y_train==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
# KNeighborsClassifier() 今回は不要の説明が表示される
# predict(X) test入れない
n_neighbors_list = [1, 3, 5]
y_true = y_test

for n_neighbors in n_neighbors_list:
    neigh = KNeighborsClassifier(n_neighbors)
    neigh.fit(standarization_X_train, y_train)
    y_pred = neigh.predict(standarization_X_test)
    
    #print('{}n-n, train推定{}'.format(n_neighbors,neigh.predict(standarization_X_train)))
    print('{}n-n, test推定{}'.format(n_neighbors,y_pred))
    # accuracyの計算
    # best score with noramalize = True is 1
    accuracy = accuracy_score(y_true, y_pred, normalize=True)
    print('accuracy{}'.format(accuracy))
    # precision macro_f
    precision_ma = precision_score(y_true, y_pred, average='macro')
    print('precision_macrof{}'.format(precision_ma))
    # precision micro_f
    precision_mi = precision_score(y_true, y_pred, average='micro')
    print('precision_microf{}'.format(precision_mi))
    # recall macro_f
    recall_ma = recall_score(y_true, y_pred, average='macro')
    print('recall_macrof{}'.format(recall_ma))
    # recall micro_f
    recall_mi = recall_score(y_true, y_pred, average='micro')
    print('recall_microf{}'.format(recall_mi))
    # confusion_matrix
    print(y_true.shape, y_pred.shape)
    matrix_of_confusion = confusion_matrix(y_true, y_pred)
    print(matrix_of_confusion)
    
    decision_region(standarization_X_train, y_train, neigh)
```

### accuracy, precision, recall, f-measure
### accuracy: 正しい予測の割合（デフォルト）またはカウント（normalize = False）
accuracy(y_true, y_pred) = 1/n Σ1*(y_true==y_pred)
予測したラベルと正解のラベルが等しい数を全体の数で割る.
### precision:  陰性のサンプルを陽性として分類しない分類器の能力
tp / (tp + fp) tp: true positive, fp: false positive(negativeをpositiveと分類)
### recall: すべての陽性サンプルを見つける分類器の能力
tp / (tp + fn) tp: true positive, fn: false negative(positiveをnegativeと分類)
### f-measure(Fβ and F1 measures): weighted harmonic mean of the precision and recall(加重調和平均)
f1 = 2 * (precision * recall) / (precision + recall)
multi class classifierはaverage = macro, micro, weighted, samplesを引数にとる
