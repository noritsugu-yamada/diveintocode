## Task7 Relationship between feature quantities

```python
#縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
#全ての組み合わせの散布図を一度に表示させる、散布図行列を作る

#df = sns.load_dataset("iris")
sns.pairplot(df, hue="Species", height=2.5)# separate colors from species
plt.show()
```

```python
df.iloc[:,:-1].transpose()
```

```python
df_corr = df.copy().corr()
sns.heatmap(df_corr,annot=True, cmap='coolwarm',)
```

```python
#Create a correlation coefficient matrix for four feature quantities
# N 行 M 列を M 行 N 列に変換して相関行列を計算する
correlation_matrix = np.corrcoef(df.iloc[:,:-1].transpose())
column_labels = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
# 相関行列のヒートマップを描く
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',
            xticklabels=column_labels,
            yticklabels=column_labels)
```

### Simple consideration from scatter plot
setosaはほぼすべての散布図で他の二つとデータのdistanceが離れており,分類するためのラインを容易に引くことができると思われる.
versicolorとversinicaはデータのdistanceが近いが,'sepal_length'と'petal_length', 'petal_length'と'sepal_width', 'petal_length'と'petal_width'
の散布図では分離可能ではないかと考えられる.
