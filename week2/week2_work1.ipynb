# 課題 統計基礎用語の説明

## 標準偏差(standard diviation)
標準偏差は分散(variance)の平方根(sqrt)で定義される
分散は分布(distribution)の平均(mean)からの距離の2乗(square)で定義され、
分散、標準偏差ともに平均からのばらつき度合を知りたい時に重要となる.
例えばN個のデータセットX(x1, x2,...,xn)の集合の場合に平均は１/NΣXnとなり,これをμとすると分散はΣ(xn-μ)^2となりこれをVとすると標準偏差は√Vとなる.
標準偏差と分散の違いについて最も大事なことは測定単位(Unit of measurement)が異なることで
ある.分散は二乗しているためグラフ化したときにx軸上に表すことができないが標準偏差は
データ点や平均と単位が同一なのでx軸上に表現することができる.

## 共分散行列

2変数の間の関係を表す数値を知りたい場合に共分散(covariance)を使う.
X1からXnまでの確率変数をまとめた集合X(X1,X2,X3,...,Xn)としたとき
それぞれの変数どうしの共分散を行列にしたものを共分散行列という
X1とX2の共分散はX1の平均をμx1,X2の平均をμx2とすると
Cov(X1,X2) = E[(X1 - μx1)(X2 - μx2)]
で定義される(E[]は平均をとるの意味).
Xの分散共分散行列をV[X]とするとV[X]は対角成分に確率変数の分散,
非対角成分にそれぞれの対応する確率変数同士の共分散を要素にもつ対称行列
となる (V[X]はXの分散).

## 1次元の正規分布と多次元の正規分布について


連続確率変数を横軸にその確率変数に対応した確率密度(連続確率変数の場合確率を定義できないため確率密度という)を表した関数を確率密度関数という.
1次元の正規分布は平均μ,分散σ2によって形が決定するパラメトリックな分布で
その確率密度関数はN(x|μ,σ2) = 1 / √2πσ2 * exp(-(x - μ)^2/2σ2)で表される.
多次元の正規分布は1次元の正規分布を多変量(多数の変数)に拡張したものと捉えられ、
その確率密度関数はNn(x|μ,Σ)= (1/√2π)^n |Σ|^-1/2 * exp(-1/2(x-μ)TΣ^-1(x-μ))
 である.|Σ|は共分散行列の行列式,Σ^-1は共分散行列の逆行列
正規分布は例えば乱数を発生させたり、観測されたデータは真の分布に少しノイズが入って観測されると仮定し、そのノイズがガウス分布に従うとしてモデル化するときに使用する.

# 課題 データ点の作成

```python
# 2次元正規分布μ=(-3,0) 共分散行列が[[1, 0.8], [0.8, 1]]のデータ生成
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
mu1 = [-3, 0]
sigma = [[1, 0.8], [0.8, 1]]
data1 = np.random.multivariate_normal(mu1, sigma, 500)
```

```python
# 2次元正規分布μ=(0,-3) 共分散行列が[[1, 0.8], [0.8, 1]]のデータ生成
mu2 = [0, -3]
data2 = np.random.multivariate_normal(mu2, sigma, 500)
data2
```

## 課題　データ点へのラベル付け

```python
# 0と1の500個の1次元配列を作る
label_0 = np.zeros(500,dtype=int)
label_1 = np.ones(500, dtype=int)
# data1にlabel_0 data2にlabel_1をつける
data1 = np.insert(data1, 2, 0, axis=1)
data2 = np.insert(data2, 2, 1, axis=1)
# data1とdata2を結合するnp.concatenate
data = np.concatenate([data1, data2], axis=0)
data
```

```python
data[:500, :1]
```

## 課題 正しくラベル付けされているか見てみる

```python
# 自分ではdataを分けないとできませんでした
import matplotlib.pyplot as plt
%matplotlib inline
x1 = data[:500, :1]
y1 = data[:500, 1:2]
x2 = data[500:, :1]
y2 = data[500:, 1:2]

plt.scatter(x1,y1, s=50, c='yellow', marker='o', label='data1')
plt.scatter(x2,y2, s=50, c='purple', marker='o', label='data2')
plt.legend(loc='upper left')
```

```python
# afterfixed c=label 
plt.title('data sample by Normal Distribution')
plt.scatter(data[:,0], data[:,1],c=data[:,2], s=50, marker='o')
plt.show()
```

```python
# Specify cmap to change the color
plt.title('data sample by Normal Distribution')
plt.ylabel('y')
plt.xlabel('x')
plt.scatter(data[:,0], data[:,1],c=data[:,2], s=50, marker='s',cmap='tab10')

plt.show()
```
